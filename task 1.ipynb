## Data Cleaning

```python
df.isnull().sum()
```

```python
df[df.Age.isnull()].sample(10)
```

```python
n=df.groupby(['Sex', 'Pclass'])[['Age']].median()
```

```python
n.Age
```

```python
df['Age']=df.groupby(['Sex', 'Pclass'],group_keys=False)['Age'].apply(lambda x: x.fillna(x.median()))
```

```python
df.Age.isnull().sum()
```

```python
df[df.Embarked.isnull()]
```

```python
df.Embarked.mode()
```

```python
df['Embarked'].fillna(df.Embarked.mode()[0],inplace=True)
```

```python
df.Embarked.isnull().sum()
```

```python
df.Cabin.isnull().sum()
```

```python
df['deck']=df['Cabin'].apply(lambda s: s[0] if pd.notnull(s) else 'M')
```

```python
df.deck.isnull().sum()
```

```python
#Cabin is dropped because Deck feature is used instead of it
df.drop(['Cabin'], inplace=True, axis=1)
```
